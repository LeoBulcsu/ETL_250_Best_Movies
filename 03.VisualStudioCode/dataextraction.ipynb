### IMDB 250 Ranking
First I need to retrieve the info that I want to use form IMDBs top 250 list
I would like to get the title, relese year, director, casting, image url and IMDB score from ech movie of the list. 

For that I am going to do some web scrapping, let's get to it:

```python
imdb_url = 'https://www.imdb.com/chart/top/'


driver.get(imdb_url)
```

I am going to need everysingle one of the movie links to get the info I want so I use selenium to find the links form the list.

```python
movies = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul').find_elements(By.XPATH, '//a[@class="ipc-title-link-wrapper"]')
```

```python
link_list=[]
for movie in movies:
    link_list.append(movie.get_attribute('href'))
```

```python
len(link_list)
```

After finding those links I need to clean a bit those extra links that Selenium found.

```python
link_list = link_list[:-8]
```

Once I have my links it is a matter of looping through each of them and extract the info I require.

```python
dict_list_IMDB = []

for link in link_list:

    driver.get(link)

    time.sleep(15)
    try:

        name = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1').text
        year = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a').text
        parental = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a').text
        length = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[3]').text
        director = driver.find_element(By.XPATH, '//main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li/a').text 
        cast = driver.find_element(By.XPATH, '//main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]').text
        cast = cast.replace('\n',' , ')
        url = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[3]/div[1]/div[1]/div/div[1]/img').get_attribute('src')
        rating = driver.find_element(By.XPATH, '//div[@class="sc-bde20123-2 gYgHoj"]').text
        rating = rating.replace('\n', '') 
        plot = driver.find_element(By.XPATH, '//span[@class="sc-466bb6c-2 eVLpWt"]').text


        tmp = {'Title': name, 'Release Year': year, 'Length': length, 'Parental': parental,'Director': director,'Cast': cast, 'IMDB rating': rating, 'Plot': plot,'Image URL': url}

        #print (link)

        dict_list_IMDB.append(tmp)
    
    except:
        try:
            name = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1').text
            year = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a').text
            parental = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a').text
            length = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[3]').text
            director = driver.find_element(By.XPATH, '//main/div/section[1]/div/section/div/div[1]/section[3]/ul/li[1]/div/ul/li/a').text
            cast = driver.find_element(By.XPATH, '//main/div/section[1]/div/section/div/div[1]/section[3]/div[2]/div[2]').text
            cast = cast.replace('\n',' as ')
            url = driver.find_element(By.XPATH, '//main/div/section[1]/section/div[3]/section/section/div[3]/div[1]/div/div/div[1]/img').get_attribute('src')
            rating = driver.find_element(By.XPATH, '//div[@class="sc-bde20123-2 gYgHoj"]').text
            rating = rating.replace('\n', '') 
            plot = driver.find_element(By.XPATH, '//span[@class="sc-466bb6c-2 eVLpWt"]').text
            
            tmp = {'Title': name, 'Release Year': year, 'Length': length, 'Parental': parental,'Director': director,'Cast': cast, 'IMDB rating': rating, 'Plot': plot,'Image URL': url}

            dict_list_IMDB.append(tmp)

        except:
            print(f'I can not do this {link}')
```
