```python
import pandas as pd
import selenium
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import time

import warnings
warnings.filterwarnings("ignore")
```

```python
imdb = pd.read_csv('../02.data_csv/IMDB.csv')
imdb_df = pd.DataFrame(imdb)
imdb_df.head()
```

```python
imdbTitleYear = imdb_df[['Title','Release Year']].values.tolist()
```

```python
driver = webdriver.Firefox()
```

```python
tomato_data_2 = []

for movie in imdbTitleYear[:17]:
    
    finder = driver.get('https://www.google.com')

    try:

        cookies = driver.find_element(By.XPATH, '//*[@id="W0wltc"]')
        cookies.click()

        time.sleep(7)

        search_bar = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
        search_bar.click()

        search_bar.send_keys(f'Rotten Tomatoes {movie}')
        search_bar.send_keys(Keys.ENTER)

        time.sleep(3)

        driver.find_element(By.XPATH, '//div[@class="byrV5b"]').click()

        time.sleep(5)

        driver.find_element(By.XPATH, '//*[@id="onetrust-reject-all-handler"]').click()

        data = driver.find_element(By.CSS_SELECTOR, '#scoreboard').text.split('\n')[1:]

        tomato_data_2.append([movie, data])

    except:
        try:
            search_bar = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
            search_bar.click()

            search_bar.send_keys(f'Rotten Tomatoes {movie}', Keys.ENTER)
            #search_bar.send_keys(Keys.ENTER)

            time.sleep(8)

            driver.find_element(By.XPATH, '//div[@class="byrV5b"]').click()

            time.sleep(8)

            data = driver.find_element(By.CSS_SELECTOR, '#scoreboard').text.split('\n')[1:]

            tomato_data_2.append([movie, data])

        except:
            print(f'I can not this {movie}')
```

```python
rottenTomatoes['Title'] = rottenTomatoes['Title'].apply(lambda x: x[0])
```

```python
rottenTomatoes
```

```python

def scores(column):
    
    tomatometer_audience = []
   
        
    if len(column) == 4:
        
        tomatometer = column[0]
        audience_score = column[2]
        tomatometer_audience.append(tomatometer)
        tomatometer_audience.append(audience_score)
    
    elif len(column) == 3:

        tomatometer = column[1]
        audience_score = column[1]
        tomatometer_audience.append(tomatometer)
        tomatometer_audience.append(audience_score)

    else:
        pass
    

    return tomatometer_audience
```

```python
rottenTomatoes['Score'] = rottenTomatoes['Score'].apply(scores)
```

```python
rottenTomatoes[['Tomatometer', 'Audience_Score']] = rottenTomatoes['Score'].apply(lambda x: pd.Series([x[0], x[1]]))
```

```python
rottenTomatoes = rottenTomatoes.drop('Score', axis=1)
```
